```python
# 2. Checking for mixed data and duplicates

df.shape
```

```python
for col in df.columns.tolist():
    weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df[weird]) > 0:
        print (col)
```

```python
# seeing the types of data in each column
df.head(50)
```

```python
#checking values in the neighbourhood_group columnn
df['neighbourhood_group'].value_counts(dropna = False)
```

```python
# Changing name column into string data type
df['name'] = df['name'].astype('str')
```

```python
# Changing host_name column into string data type
df['host_name'] = df['host_name'].astype('str')
```

```python
# Changing neighbourhood_group column into string data tpye
df['neighbourhood_group'] = df['neighbourhood_group'].astype('str')
```

```python
# Changing last_review column into dataime64 data type
df['last_review'] = df['last_review'].astype('datetime64')
```

```python
# Checking for duplicates
df_dups = df[df.duplicated()]
```

```python
df_dups
```

## 3. Dropping Columns

```python
# Removing host_name for PII reasons
df = df.drop(columns = ['host_name'])
```

```python
df.head()
```

```python
# Checking min and Max values of columns to remove outliers
df.describe()
```

```python
# Removing max display rows in pandas output
pd.set_option('display.max_rows', None)
```

```python
# Checking counts of minimum_nights
df['minimum_nights'].value_counts(dropna = False).sort_index()
```

```python
# Creating new dataset that removes all rows that have minimum_nights greater than or equal to 366
df2 = df.loc[df['minimum_nights'] <= 366]
```

```python
# Creating new dataset that removes all rows that have a price greater than 9999
df3 = df2.loc[df2['price'] <= 9999]
```
